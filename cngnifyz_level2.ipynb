**Task: Restaurant Chains**

* Identify if there are any restaurant chains
present in the dataset.

* Analyze the ratings and popularity of
different restaurant chains.

```python
# Count occurrences of each restaurant name
chain_counts = df['Restaurant Name'].value_counts()

# Keep only names appearing more than once
chains = chain_counts[chain_counts > 1]

print("Top Restaurant Chains:")
print(chains.head(10))
```

```python
# Filter only chain restaurants
chain_df = df[df['Restaurant Name'].isin(chains.index)]
```

```python
rated_chain_df = chain_df[chain_df['Aggregate rating'] > 0]

chain_analysis = rated_chain_df.groupby('Restaurant Name').agg(
    number_of_outlets=('Restaurant ID', 'count'),
    avg_rating=('Aggregate rating', 'mean'),
    avg_votes=('Votes', 'mean')
)

# Sort by number of outlets
chain_analysis = chain_analysis.sort_values(by='number_of_outlets', ascending=False)

chain_analysis.head(10)

```

```python
popular_chains = chain_analysis.sort_values(by='avg_votes', ascending=False)

popular_chains.head(10)
```

```python
strong_chains = chain_analysis[chain_analysis['number_of_outlets'] >= 5]

top_rated_chains = strong_chains.sort_values(by='avg_rating', ascending=False)

top_rated_chains.head(10)
```

```python
chain_analysis.head(10)['number_of_outlets'].plot(kind='bar')

import matplotlib.pyplot as plt
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.ylabel("Number of Outlets")
plt.show()
```
